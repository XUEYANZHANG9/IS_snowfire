```python
################## INPUTS: BASIN AND MODEL (OR ENSEMBLE AVERAGE OR MEDIAN) ################################################

import matplotlib
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib.pyplot as plot
from snowpack_functions import make_map 
plt.ion()
scenario = "historical"
# basin = "california"
basins = ["california","cascades","northernrockies","southernrockies","whites"] 

fig = plt.figure(figsize=(15,15))
m = make_map()
for basin in basins: 

    ##### add points
    ### load lats and lons
    file = '/raid9/gergel/agg_snowpack/%s/%s.npz' % (scenario,basin)
    data = np.load(file)
    lats_lons = data['latslons']
    lats = lats_lons[:,0]
    lons = lats_lons[:,1]
    
    x,y = m(lons,lats)
    ms = 0.5
    m.scatter(x,y,marker='.',color = 'b',s=ms)

# plt.show()

### save plot
import os
direc = '/raid9/gergel/agg_snowpack/maps/'
model_spec = 'map'
plotname = 'historical_domain_map' 
savepath = os.path.join(direc,plotname)
print ("saving figure to '%s'" % savepath)
plt.savefig(savepath)

```
